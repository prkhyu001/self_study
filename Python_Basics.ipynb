````python
#간단한 계산기
cost = input("상품의 가격 : ")
payment = input("지불한 금액 : ")
#change = cost - paymet >>> input()은 문자열로 입력 받기에 연산 계산 불가능 >>>에러
change = int(payment) - int(cost)
#print("나머지 금액은 " + change + "원 입니다.") 문자열은 문자열끼리 
#사용해야 하기에, 에러가 뜹니다.
print("상품의 가격은" + cost + "원 이며, 잔돈은" + str(change) + "원 입니다.")
````
