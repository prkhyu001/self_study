# 5가지 다른 방법으로 양수 배열의 정수 부분 추출하기

```python
# 10개의 정수를 생성하고 이를 사용해 배열을 만드는 함수 
def generate():
    for x in range(10):
        yield x

z = np.fromiter(generate(), dtype=float, count = -1)
print(z)
```
